````python
import matplotlib.pyplot as plt
from google.colab import files

category_revenue = df.groupby('Category')['Total_Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))

category_revenue.plot(
    kind='bar',
    color='royalblue',
    edgecolor='black'
)

plt.title("Revenue by Product Category", fontsize=16, fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Revenue (₹)")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()

plt.savefig(
    "revenue_by_category.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()

files.download("revenue_by_category.png")
````

````python
import matplotlib.pyplot as plt
from google.colab import files

monthly_revenue = df.groupby('Month')['Total_Amount'].sum()

# Arrange months in calendar order
month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

monthly_revenue = monthly_revenue.reindex(month_order)

plt.figure(figsize=(10,5))

monthly_revenue.plot(
    kind='line',
    marker='o',
    linewidth=3,
    color='royalblue'
)

plt.title("Monthly Revenue Trend", fontsize=16, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Revenue (₹)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()

plt.savefig(
    "monthly_revenue_trend.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()

files.download("monthly_revenue_trend.png")
````

````python
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

corr = df[['Age','Quantity','Unit_Price','Total_Amount']].corr()

fig, ax = plt.subplots(figsize=(6,5))

im = ax.imshow(corr, cmap='coolwarm')

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))

ax.set_xticklabels(corr.columns, rotation=45)
ax.set_yticklabels(corr.columns)

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        ax.text(
            j,
            i,
            f"{corr.iloc[i,j]:.2f}",
            ha="center",
            va="center",
            color="black"
        )

plt.title("Correlation Heatmap")

plt.colorbar(im)

plt.tight_layout()

plt.savefig(
    "correlation_heatmap.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()

files.download("correlation_heatmap.png")
````

Correlation Analysis:
The correlation matrix shows that Quantity (0.651) and Unit Price (0.688) have moderate to strong positive correlations with Total Amount, indicating that both purchasing more items and buying higher-priced products increase overall revenue. Age has almost no correlation with Quantity, Unit Price, or Total Amount, suggesting that customer spending behavior is not significantly influenced by age in this dataset.
